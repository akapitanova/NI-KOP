```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

from cProfile import label

%matplotlib inline
```

```python
pd.set_option('display.max_rows', None)
```

```python
path_gsat_tezke_ins = 'gsat201-gsat204-91.csv'
path_gsat_lehke_ins = 'gsat201-gsat204-71.csv'
```

```python
gsat_lehke_ins = pd.read_csv(path_gsat_lehke_ins, delimiter = ';')
gsat_tezke_ins = pd.read_csv(path_gsat_tezke_ins, delimiter = ';')
```

```python
gsat_lehke_ins
```

```python
gsat_probsat_lehke_ins = gsat_lehke_ins[['inst', 'succ B', 'of', 'steps B', 'awg fined B', 'mu B', 'sig^2 B']]
gsat_probsat_lehke_ins.rename(columns={'succ B' :'succ gsat', 'steps B' :'steps gsat', 
                                       'awg fined B' :'awg fined gsat', 'mu B' :'mu gsat',
                                       'sig^2 B' :'sig^2 gsat'},inplace=True)
gsat_probsat_tezke_ins = gsat_tezke_ins[['inst', 'succ B', 'of', 'steps B', 'awg fined B', 'mu B', 'sig^2 B']]
gsat_probsat_tezke_ins.rename(columns={'succ B' :'succ gsat', 'steps B' :'steps gsat', 
9                                       'awg fined B' :'awg fined gsat', 'mu B' :'mu gsat',
                                       'sig^2 B' :'sig^2 gsat'},inplace=True)
```

```python
gsat_probsat_lehke_ins
```

```python
directory_71 = 'results_71'
directory_91 = 'results_91'
```

```python
gsat_probsat_lehke_ins['succ probsat'] = np.nan
```

```python
def fill_df(df_to_fill, directory):
    for filename in os.scandir(directory):
        if filename.is_file():
            df = pd.read_csv(filename.path, delimiter=' ',names=['iter', 'max_iter','sat_clauses','max_sat_clauses'])
            df['iter'] += 1

            # succ probsat
            df_to_fill.loc[df_to_fill['inst'] == filename.name.split('.')[0], 
                   'succ probsat'] = len(df[df['iter'] != df['max_iter']])

            # steps probsat
            df_to_fill.loc[df_to_fill['inst'] == filename.name.split('.')[0], 
                   'steps probsat'] = df['iter'].mean()

            # awg fined probsat
            n = len(df)
            sm = 0
            for i in range(n):
                if df['iter'][i] == df['max_iter'][i]:
                    sm += 10 * df['iter'][i]
                else:
                    sm += df['iter'][i]
            fined_awg = sm / n
            df_to_fill.loc[df_to_fill['inst'] == filename.name.split('.')[0], 
                   'awg fined probsat'] = fined_awg

            # mu probsat
            iters = np.log(df['iter'])
            mu = sum(iters) / n
            df_to_fill.loc[df_to_fill['inst'] == filename.name.split('.')[0], 
                   'mu probsat'] = mu

            # sig^2 probsat
            sigma2 = np.sum([(x_i - mu) ** 2 for x_i in iters]) / (n - 1)
            df_to_fill.loc[df_to_fill['inst'] == filename.name.split('.')[0], 
                   'sig^2 probsat'] = sigma2
```

```python
fill_df(gsat_probsat_lehke_ins, directory_71)
```

```python
fill_df(gsat_probsat_tezke_ins, directory_91)
```

```python
def make_graph(feature_gsat, feature_probsat):
    bins = 80
    plt.figure(figsize=(15, 15))

    # histogram of the data
    plt.subplot(2, 2, 1)
    gsat_probsat_lehke_ins[feature_gsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', density = True)
    plt.title("Alg gsat lehke")

    # histogram of the data
    plt.subplot(2, 2, 2)
    gsat_probsat_lehke_ins[feature_probsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', density = True)
    plt.title("Alg probsat lehke")

    # histogram of the data
    plt.subplot(2, 2, 3)
    gsat_probsat_tezke_ins[feature_gsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', density = True)
    plt.title("Alg gsat tezke")

    # histogram of the data
    plt.subplot(2, 2, 4)
    gsat_probsat_tezke_ins[feature_probsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', density = True)
    plt.title("Alg probsat tezke")

    plt.show()
```

```python
make_graph('steps gsat', 'steps probsat')
```

```python
make_graph('awg fined gsat', 'awg fined probsat')
```

```python
make_graph('succ gsat', 'succ probsat')
```
