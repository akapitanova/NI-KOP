```python
# filter out runs where no solution was found (sat_clauses != max_sat_clauses)
def filter_data(df):
    return df[df['sat_clauses'] == df['max_sat_clauses']]
```

```python
# filtrovani dat
stat1000_0_1 = pd.read_csv(path_stat1000_0_1, delimiter=' ',names=['iter', 'max_iter','sat_clauses','max_sat_clauses'])
stat1000_0_1_succ = filter_data(stat1000_0_1)
stat1000_0_4 = pd.read_csv(path_stat1000_0_4, delimiter=' ',names=['iter', 'max_iter','sat_clauses','max_sat_clauses'])
stat1000_0_4_succ = filter_data(stat1000_0_4)
```

### Histogram pro prob 0.4 na uspesnych

```python
iters = np.log(stat1000_0_4_succ['iter'])
mean = iters. / len(iters)
print(mean)
```

```python
bins = 40

# historgram of the data
stat1000_0_4_succ['iter'].plot(kind = 'hist', bins = bins, edgecolor = 'black', density = True)
plt.show()

# histogram of logarithm of the data 
# (if the data has lognormal distribution, 
# the logarithm of the data should be normally distributed )
np.log(stat1000_0_4_succ['iter']).plot(kind = 'hist',bins = bins, edgecolor = 'black', density = True)

# estimation of parameters of the lognormal distribution
iters = np.log(stat1000_0_4_succ['iter'])
mean = sum(iters) / len(iters)
stddev = np.sum([(x_i - mean) ** 2 for x_i in iters]) / len(iters)

# ploting graph of normal distribution with the estimated parameters
mu = mean
variance = stddev
sigma = np.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()

print(f'Estimated parameters: mu = {mu}, sigma = {sigma}')
```

### Srovnani pro CDF na uspesnych

```python
def ecdf(data):
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n + 1) / n
    return x,y
```

```python
x,y = ecdf(stat1000_0_1_succ['iter'])
plt.plot(x,y,label='0.1')
x,y = ecdf(stat1000_0_4_succ['iter'])
plt.plot(x,y,label='0.4')
plt.legend()
plt.show()

print("Celkove z 1000 je s:")
print("Prob 0.1 uspesnych : " + str(len(stat1000_0_1_succ)))
print("Prob 0.4 uspesnych : " + str(len(stat1000_0_4_succ)))
```

### Srovnani pro CDF na vsech

```python
def ecdf_all(data, data_all):
    n_succ = data.size
    x = np.sort(data)
    n_all = data_all.size
    y = np.arange(1, n_succ + 1) / n_succ
    y = y * (n_succ/n_all)
    return x,y
```

```python
x,y = ecdf_all(stat1000_0_1_succ['iter'], stat1000_0_1['iter'])
plt.plot(x,y,label='0.1')
x,y = ecdf_all(stat1000_0_4_succ['iter'], stat1000_0_4['iter'])
plt.plot(x,y,label='0.4')
plt.legend()
plt.show()
```
