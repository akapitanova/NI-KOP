```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = 'konverg.dat'
df = pd.read_csv(file,delimiter=' ',names=['iter', 'sat_clauses'])
df = df.set_index('iter')
df.hist(bins = 20,edgecolor='black')
plt.show()
```

```python
# filter out runs where no solution was found (sat_clauses != max_sat_clauses)
def filter_data(df):
    return df[df['sat_clauses'] == df['max_sat_clauses']]
```

```python
file = '~/bin/stat.dat'
bins = 40

df2 = pd.read_csv(file,delimiter=' ',names=['iter', 'max_iter','sat_clauses','max_sat_clauses'])
df2 = filter_data(df2)
# historgram of the data
df2['iter'].plot(kind='hist', bins=bins,edgecolor='black',density=True)
plt.show()
# histogram of logarithm of the data (if the data has lognormal distribution, the logarithm of the data should be normally distributed )
np.log(df2['iter']).plot(kind='hist',bins=bins,edgecolor='black',density = True)

# estimation of parameters of the lognormal distribution
iter = np.log(df2['iter'])
mean = iter.sum()/len(iter)
stddev = np.sum([(x_i - mean) ** 2 for x_i in iter])/len(iter)

# ploting graph of normal distribution with the estimated parameters
import scipy.stats as stats
mu = mean
variance = stddev
sigma = np.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()

print(f'Estimated parameters: mu = {mu}, sigma = {sigma}')
```

```python
def ecdf(data):
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return x,y
```

```python
from cProfile import label


df_0_1 = pd.read_csv('~/bin/stat_0_1.dat',delimiter=' ',names=['iter', 'max_iter','sat_clauses','max_sat_clauses'])
df_0_1 = filter_data(df_0_1)
df_0_4 = pd.read_csv('~/bin/stat_0_4.dat',delimiter=' ',names=['iter', 'max_iter','sat_clauses','max_sat_clauses'])
df_0_4 = filter_data(df_0_4)

x,y = ecdf(df_0_1['iter'])
plt.plot(x,y,label='0.1')
x,y = ecdf(df_0_4['iter'])
plt.plot(x,y,label='0.4')
plt.legend()
plt.show()
```
