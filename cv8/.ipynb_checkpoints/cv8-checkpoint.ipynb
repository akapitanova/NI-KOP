```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

%matplotlib inline
```

```python
pd.set_option('display.max_rows', None)
```

```python
path_20_095 = 'res_20_095.csv'
path_20_080 = 'res_20_080.csv'

path_10_095 = 'res_10_095.csv'
path_10_080 = 'res_10_080.csv'
```

```python
def make_graph(feature, df1=df_20_095, df2=df_20_080, titledf1='df_20_095', titledf2='df_20_080', HEIGTH=12, WIDTH=12, ROWS=1, COLS=1):
    bins = 30
    plt.figure(figsize=(WIDTH, HEIGTH))
    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    df1[feature].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='b', density = True, alpha=0.5)
    

    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    df2[feature].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='y', density = True, alpha=0.5)
    plt.legend(labels=[titledf1, titledf2])
    plt.title("Tezke instance")

    plt.show()
```
