# Porovnání algoritmů

```python
def make_graph(feature_gsat, feature_probsat, HEIGTH, WIDTH, ROWS, COLS):
    bins = 100
    plt.figure(figsize=(WIDTH, HEIGTH))

    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    gsat_probsat_lehke_ins[feature_gsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='b', density = True, alpha=0.5)

    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    gsat_probsat_lehke_ins[feature_probsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='y', density = True, alpha=0.5)
    plt.legend(labels=['gsat', 'probsat'])
    plt.title("Lehke instance")

    # histogram of the data
    plt.subplot(ROWS, COLS, 2)
    gsat_probsat_tezke_ins[feature_gsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='b', density = True, alpha=0.5)
    

    # histogram of the data
    plt.subplot(ROWS, COLS, 2)
    gsat_probsat_tezke_ins[feature_probsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='y', density = True, alpha=0.5)
    plt.legend(labels=['gsat', 'probsat'])
    plt.title("Tezke instance")

    plt.show()
```

```python
def make_graph_1(feature_gsat, feature_probsat, HEIGTH, WIDTH, ROWS, COLS):
    bins = 100
    plt.figure(figsize=(WIDTH, HEIGTH))
    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    gsat_probsat_tezke_ins[feature_gsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='b', density = True, alpha=0.5)
    

    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    gsat_probsat_tezke_ins[feature_probsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='y', density = True, alpha=0.5)
    plt.legend(labels=['gsat', 'probsat'])
    plt.title("Tezke instance")

    plt.show()
```

```python
def make_graph_2(feature_gsat, feature_probsat, HEIGTH, WIDTH, ROWS, COLS):
    bins = 100
    plt.figure(figsize=(WIDTH, HEIGTH))
    # histogram of the data
    plt.subplot(ROWS, COLS, 1)
    gsat_probsat_lehke_ins[feature_gsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='b', density = True, alpha=0.5)
    plt.legend(labels=['gsat'])

    # histogram of the data
    plt.subplot(ROWS, COLS, 2)
    gsat_probsat_lehke_ins[feature_probsat].plot(kind = 'hist', bins = bins, edgecolor = 'black', color='y', density = True, alpha=0.5)
    plt.legend(labels=['probsat'])

    plt.suptitle("Lehke instance vazeny prumer kroku", fontsize=16)
    plt.show()
```

# Srovnání průměrného počtu kroků algoritmů
